```python
import numpy as np
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Matrix multiplication
print(np.dot(A, B))

# Transpose
print(A.T)

# Determinant
print(np.linalg.det(A))
```

```python
A = np.array([[2, 1],
              [1, 1]])

B = np.array([5, 3])

solution = np.linalg.solve(A, B)
print(solution)
```

```python
import time

large = np.random.rand(10_000_000)

# Python loop
start = time.time()
total = 0
for i in large:
    total += i
end = time.time()
print("Loop time:", end - start)

# NumPy
start = time.time()
total_np = np.sum(large)
end = time.time()
print("NumPy time:", end - start)
```
