📌 1. NumPy Broadcasting
🔍 Definition

Broadcasting is a mechanism that allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array without copying data.

⚙️ Broadcasting Rules

Two arrays are compatible if:

Their dimensions are equal OR

One of them is 1

Comparison happens from right to left (trailing dimensions).

✅ Example 1 — 1D + 2D Array

```python
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([10, 20, 30])

print(A + B)
```

💡 What happens internally:

A.shape = (2, 3)

B.shape = (3,) → treated as (1,3)

NumPy expands B to:

```python
[[10, 20, 30],
 [10, 20, 30]]
```

✅ Output:
[[11 22 33]
 [14 25 36]]

✅ Example 2 — Scalar Broadcasting

```python
print(A * 2)
```

👉 Scalar 2 is applied to every element:

[[ 2  4  6]
 [ 8 10 12]]
✅ Example 3 — Column Broadcasting

```python
C = np.array([[1],
              [2]])

print(A + C)
```

Shapes:

A → (2,3)

C → (2,1)

C expands across columns:

[[1,1,1],
 [2,2,2]]
Output:
[[2 3 4]
 [6 7 8]]
🧠 Key Insight

Broadcasting avoids:

Writing loops

Creating unnecessary copies
→ Memory efficient + faster

⚡ 2. NumPy Vectorisation
🔍 Definition

Vectorisation means performing operations on entire arrays at once instead of iterating element-by-element using loops.

❌ Without Vectorisation (Slow)

```python
arr = np.array([1, 2, 3, 4, 5])

result = []
for x in arr:
    result.append(x ** 2)

print(result)
```

✅ With Vectorisation (Fast)

```python
result = arr ** 2
print(result)
```

👉 Entire operation happens in one step (C-level execution)

✅ Example 1 — Element-wise Operations

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a * b)
```

✅ Example 2 — Conditional Vectorisation

```python
arr = np.array([-3, -1, 2, 4])

arr[arr < 0] = 0
print(arr)
```

👉 Replaces all negatives without loops

✅ Example 3 — Normalization

```python
data = np.array([10, 20, 30, 40])

normalized = (data - np.min(data)) / (np.max(data) - np.min(data))
print(normalized)
```

⚡ Performance Comparison

```python
import time

large = np.random.rand(1_000_000)

# Loop
start = time.time()
s = 0
for x in large:
    s += x
print("Loop:", time.time() - start)

# Vectorised
start = time.time()
s = np.sum(large)
print("NumPy:", time.time() - start)
```

👉 NumPy is significantly faster because:

Uses optimized C backend

Avoids Python interpreter overhead

Uses SIMD (parallel CPU operations)

Example	(2×3) + (3,)	arr²
